```python
#Skip
#Imports
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Skip
#Data Entry
NAdf = pd.read_csv("data/DataViz Project - NATeams.csv")
EUdf = pd.read_csv("data/DataViz Project - EUTeams.csv")
NAEUdf = pd.read_csv("data/DataViz Project - NAEUTeams.csv")
```

```python
#Skip
#Styling
sns.set_style('dark')
sns.set_palette(['tab:blue', 'tab:orange'])
```

```python
#Overall Plot Creation
plt.figure(figsize=(20,10))
NAEUplot = sns.scatterplot(NAEUdf, x = 'games', y = 'win percentage', hue="region")
NAEUplot.set_title("NA & EU Win Percentage")
NAEUplot.set_ylabel("Win Percentage %")
NAEUplot.set_xlabel("Matches Played")
NAEUplot.legend(loc='upper left')



plt.show()
```

```python
baseChart, axes = plt.subplots(2, 2, figsize=(20,10))
plt.subplots_adjust(hspace = 0.3, wspace=0.3)
baseChart.suptitle("Stats VS Win Percentage")

NAEUSPGplot = sns.scatterplot(NAEUdf, x='score per game', y='win percentage', ax=axes[0, 0], hue="region")
NAEUGPGplot = sns.scatterplot(NAEUdf, x='goals per game', y='win percentage', ax=axes[0, 1], hue="region")
NAEUAPSplot = sns.scatterplot(NAEUdf, x='assists per game', y='win percentage', ax=axes[1, 0], hue="region")
NAEUHPGplot = sns.scatterplot(NAEUdf, x='saves per game', y='win percentage', ax=axes[1, 1], hue="region")


NAEUSPGplot.legend(loc='upper left')
NAEUGPGplot.legend(loc='upper left')
NAEUAPSplot.legend(loc='upper left')
NAEUHPGplot.legend(loc='upper right')


NAEUSPGplot.set_ylabel("Win Percentage %")
NAEUGPGplot.set_ylabel("Win Percentage %")
NAEUAPSplot.set_ylabel("Win Percentage %")
NAEUHPGplot.set_ylabel("Win Percentage %")


NAEUSPGplot.set_xlabel("Score AVG")
NAEUGPGplot.set_xlabel("Goals AVG")
NAEUAPSplot.set_xlabel("Assist AV")
NAEUHPGplot.set_xlabel("Saves AVG")




plt.show()

#NAEUplot = sns.scatterplot(NAdf, x = , y = 'win percentage', hue="region")
```

```python
boostChart, axes = plt.subplots(3, 2, figsize=(20,10))
plt.subplots_adjust(hspace = 0.3, wspace=0.3)
boostChart.suptitle("Boost Stats VS Win Percentage")

NAEUBPMGplot = sns.scatterplot(NAEUdf, x='bpm per game', y='win percentage', ax=axes[0, 0], hue="region")
NAEUABAPGplot = sns.scatterplot(NAEUdf, x='avg boost amount per game', y='win percentage', ax=axes[0, 1], hue="region")
NAEUASPGplot = sns.scatterplot(NAEUdf, x='amount stolen per game', y='win percentage', ax=axes[1, 0], hue="region")
NAEUZBTPGplot = sns.scatterplot(NAEUdf, x='0 boost time per game', y='win percentage', ax=axes[1, 1], hue="region")
NAEUHBTPGplot = sns.scatterplot(NAEUdf, x='100 boost time per game', y='win percentage', ax=axes[2, 0], hue="region")
NAEUAOSPGplot = sns.scatterplot(NAEUdf, x='amount overfill total per game', y='win percentage', ax=axes[2, 1], hue="region")

NAEUBPMGplot.legend(loc='upper left')
NAEUABAPGplot.legend(loc='upper left')
NAEUASPGplot.legend(loc='upper left')
NAEUZBTPGplot.legend(loc='upper left')
NAEUHBTPGplot.legend(loc='upper left')
NAEUAOSPGplot.legend(loc='upper left')

NAEUBPMGplot.set_ylabel("Win Percentage %")
NAEUABAPGplot.set_ylabel("Win Percentage %")
NAEUASPGplot.set_ylabel("Win Percentage %")
NAEUZBTPGplot.set_ylabel("Win Percentage %")
NAEUHBTPGplot.set_ylabel("Win Percentage %")
NAEUAOSPGplot.set_ylabel("Win Percentage %")

NAEUBPMGplot.set_xlabel("Boost Per Minute AVG")
NAEUABAPGplot.set_xlabel("Team Boost AVG")
NAEUASPGplot.set_xlabel("Stolen Boost AVG")
NAEUZBTPGplot.set_xlabel("Zero Boost Time AVG")
NAEUHBTPGplot.set_xlabel("Hundred Boost Time AVG")
NAEUAOSPGplot.set_xlabel("Overfill AVG")

plt.show()

```

```python
timeChart, axes = plt.subplots(3, 2, figsize=(20,10))

plt.subplots_adjust(hspace = 0.3, wspace=0.3)
timeChart.suptitle("Times VS Win Percentage")

NAEUTSPGplot = sns.scatterplot(NAEUdf, x='time slow speed per game', y='win percentage', ax=axes[0, 0], hue="region")
NAEUTBPGplot = sns.scatterplot(NAEUdf, x='time boost speed per game', y='win percentage', ax=axes[0, 1], hue="region")
NAEUTFPGplot = sns.scatterplot(NAEUdf, x='time supersonic speed per game', y='win percentage', ax=axes[1, 0], hue="region")
NAEUTOGAplot = sns.scatterplot(NAEUdf, x='time on ground per game', y='win percentage', ax=axes[1, 1], hue="region")
NAEUTOLAplot = sns.scatterplot(NAEUdf, x='time low in air per game', y='win percentage', ax=axes[2, 0], hue="region")
NAEUTOHAplot = sns.scatterplot(NAEUdf, x='time high in air per game', y='win percentage', ax=axes[2, 1], hue="region")

NAEUTSPGplot.legend(loc='upper left')
NAEUTBPGplot.legend(loc='upper left')
NAEUTFPGplot.legend(loc='upper left')
NAEUTOGAplot.legend(loc='upper left')
NAEUTOLAplot.legend(loc='upper left')
NAEUTOHAplot.legend(loc='upper left')

NAEUTSPGplot.set_ylabel("Win Percentage %")
NAEUTBPGplot.set_ylabel("Win Percentage %")
NAEUTFPGplot.set_ylabel("Win Percentage %")
NAEUTOGAplot.set_ylabel("Win Percentage %")
NAEUTOLAplot.set_ylabel("Win Percentage %")
NAEUTOHAplot.set_ylabel("Win Percentage %")

NAEUTSPGplot.set_xlabel("Boost Per Minute AVG")
NAEUTBPGplot.set_xlabel("Team Boost AVG")
NAEUTFPGplot.set_xlabel("Stolen Boost AVG")
NAEUTOGAplot.set_xlabel("0 Boost Time AVG")
NAEUTOLAplot.set_xlabel("100 Boost Time AVG")
NAEUTOHAplot.set_xlabel("Overfill AVG")

plt.show()
```
