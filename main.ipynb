```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib widget
```

```python
totaldf = pd.read_csv("data/DataViz Project - NAEUTeams.csv")
print(totaldf)
```

```python
sb.scatterplot(totaldf, x='games', y='win percentage', hue='region')

teamName = totaldf['team name']
allGames = totaldf['games']
allPercents = totaldf['win percentage']



for i, team in enumerate(teamName):
    plt.text(allGames[i] + 0.2, allPercents[i] + 0.2, team)

plt.show()
```
