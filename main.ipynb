```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mp
import seaborn as sb
import dash
```

```python
nateamdf = pd.read_csv("data/DataViz Project - NATeams.csv")
euteamdf = pd.read_csv("data/DataViz Project - EUTeams.csv")
totaldf = nateamdf.append(euteamdf)
print(totaldf)
```

```python
sb.scatterplot(data=totaldf, x='games', y='wins', hue='region')
plt.xlabel = "Games"
plt.ylabel = "Wins"
plt.title = "Games VS Wins"

plt.show()
```
