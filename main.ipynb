```python
#Skip
#Imports
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Skip
#Data Entry
NAdf = pd.read_csv("data/DataViz Project - NATeams.csv")
EUdf = pd.read_csv("data/DataViz Project - EUTeams.csv")
NAEUdf = pd.read_csv("data/DataViz Project - NAEUTeams.csv")
```

```python
#Skip
#Styling
sns.set_style('dark')
sns.set_palette(['tab:blue', 'tab:orange'])
```

```python
plt.figure(figsize=(10, 8))
NAEUplot = sns.scatterplot(NAEUdf, x = 'win percentage', y = 'games', hue="region", alpha=0.6, s=80)
NAEUplot.set_title("NA & EU Win Percentage")
NAEUplot.set_ylabel("Matches Played")
NAEUplot.set_xlabel("Win Percentage %")
NAEUplot.legend(loc='upper left')
plt.setp(NAEUplot.get_legend().get_texts(), fontsize='15')
plt.show()
```

```python
baseChart, axes = plt.subplots(2, 2, figsize=(20,10))
plt.subplots_adjust(hspace = 0.3, wspace=0.3)
baseChart.suptitle("Stats VS Win Percentage")

NAEUSPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='score per game',  ax=axes[0, 0], hue="region", alpha=0.6)
NAEUGPGplot = sns.scatterplot(NAEUdf,  x='win percentage',y='goals per game', ax=axes[0, 1], hue="region", alpha=0.6)
NAEUAPSplot = sns.scatterplot(NAEUdf,  x='win percentage', y='assists per game',ax=axes[1, 0], hue="region", alpha=0.6)
NAEUHPGplot = sns.scatterplot(NAEUdf,  x='win percentage', y='saves per game',  ax=axes[1, 1], hue="region", alpha=0.6)


NAEUSPGplot.legend(loc='upper left')
NAEUGPGplot.legend(loc='upper left')
NAEUAPSplot.legend(loc='upper left')
NAEUHPGplot.legend(loc='upper right')


NAEUSPGplot.set_xlabel("Win Percentage %")
NAEUGPGplot.set_xlabel("Win Percentage %")
NAEUAPSplot.set_xlabel("Win Percentage %")
NAEUHPGplot.set_xlabel("Win Percentage %")


NAEUSPGplot.set_ylabel("Score AVG")
NAEUGPGplot.set_ylabel("Goals AVG")
NAEUAPSplot.set_ylabel("Assist AVG")
NAEUHPGplot.set_ylabel("Saves AVG")




plt.show()

#NAEUplot = sns.scatterplot(NAdf, x = , y = 'win percentage', hue="region")
```

```python
boostChart, axes = plt.subplots(3, 2, figsize=(20,10))
plt.subplots_adjust(hspace = 0.3, wspace=0.3)
boostChart.suptitle("Boost Stats VS Win Percentage")

NAEUBPMGplot = sns.scatterplot(NAEUdf, x='win percentage', y='bpm per game',  ax=axes[0, 0], hue="region", alpha=0.6)
NAEUABAPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='avg boost amount per game',  ax=axes[0, 1], hue="region", alpha=0.6)
NAEUASPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='amount stolen per game',  ax=axes[1, 0], hue="region", alpha=0.6)
NAEUZBTPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='0 boost time per game',  ax=axes[1, 1], hue="region", alpha=0.6)
NAEUHBTPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='100 boost time per game',  ax=axes[2, 0], hue="region", alpha=0.6)
NAEUAOSPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='amount overfill total per game',  ax=axes[2, 1], hue="region", alpha=0.6)

NAEUBPMGplot.legend(loc='upper left')
NAEUABAPGplot.legend(loc='upper left')
NAEUASPGplot.legend(loc='upper left')
NAEUZBTPGplot.legend(loc='upper left')
NAEUHBTPGplot.legend(loc='upper left')
NAEUAOSPGplot.legend(loc='upper left')

NAEUBPMGplot.set_xlabel("Win Percentage %")
NAEUABAPGplot.set_xlabel("Win Percentage %")
NAEUASPGplot.set_xlabel("Win Percentage %")
NAEUZBTPGplot.set_xlabel("Win Percentage %")
NAEUHBTPGplot.set_xlabel("Win Percentage %")
NAEUAOSPGplot.set_xlabel("Win Percentage %")

NAEUBPMGplot.set_ylabel("Boost Per Minute AVG")
NAEUABAPGplot.set_ylabel("Team Boost AVG")
NAEUASPGplot.set_ylabel("Stolen Boost AVG")
NAEUZBTPGplot.set_ylabel("Zero Boost Time AVG")
NAEUHBTPGplot.set_ylabel("Hundred Boost Time AVG")
NAEUAOSPGplot.set_ylabel("Overfill AVG")

plt.show()

```

```python
timeChart, axes = plt.subplots(3, 2, figsize=(20,10))

plt.subplots_adjust(hspace = 0.3, wspace=0.3)
timeChart.suptitle("Times VS Win Percentage")

NAEUTSPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='time slow speed per game',  ax=axes[0, 0], hue="region", alpha=0.6)
NAEUTBPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='time boost speed per game', ax=axes[0, 1], hue="region", alpha=0.6)
NAEUTFPGplot = sns.scatterplot(NAEUdf, x='win percentage', y='time supersonic speed per game',  ax=axes[1, 0], hue="region", alpha=0.6)
NAEUTOGAplot = sns.scatterplot(NAEUdf, x='win percentage', y='time on ground per game',  ax=axes[1, 1], hue="region", alpha=0.6)
NAEUTOLAplot = sns.scatterplot(NAEUdf, x='win percentage', y='time low in air per game',  ax=axes[2, 0], hue="region", alpha=0.6)
NAEUTOHAplot = sns.scatterplot(NAEUdf, x='win percentage', y='time high in air per game',  ax=axes[2, 1], hue="region", alpha=0.6)

NAEUTSPGplot.legend(loc='upper left')
NAEUTBPGplot.legend(loc='upper left')
NAEUTFPGplot.legend(loc='upper left')
NAEUTOGAplot.legend(loc='upper left')
NAEUTOLAplot.legend(loc='upper left')
NAEUTOHAplot.legend(loc='upper left')

NAEUTSPGplot.set_xlabel("Win Percentage %")
NAEUTBPGplot.set_xlabel("Win Percentage %")
NAEUTFPGplot.set_xlabel("Win Percentage %")
NAEUTOGAplot.set_xlabel("Win Percentage %")
NAEUTOLAplot.set_xlabel("Win Percentage %")
NAEUTOHAplot.set_xlabel("Win Percentage %")

NAEUTSPGplot.set_ylabel("Time Slow Speed AVG")
NAEUTBPGplot.set_ylabel("Time Boost Speed AVG")
NAEUTFPGplot.set_ylabel("Time Supersonic AVG")
NAEUTOGAplot.set_ylabel("Time on Ground AVG")
NAEUTOLAplot.set_ylabel("Time low in air AVG")
NAEUTOHAplot.set_ylabel("Time high in air AVG")

plt.show()
```
